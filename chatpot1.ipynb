```python
import requests
import pandas as pd

# بيانات الاتصال
subscription_key = "21kxvaNtB9S64Z9SLq0QyEdFrc9waGQ5mDPi9RP9QPseDX2WsbJkJQQJ99BDACPV0roXJ3w3AAAaACOGH0pp"
endpoint = "https://dental.cognitiveservices.azure.com/"

# اسم المشروع واسم الـ deployment من Azure Language Studio
project_name = "DENATAL"
deployment_name = "production"

# URL الجديد الصحيح للـ Question Answering
url = f"{endpoint}language/:query-knowledgebases?projectName={project_name}&deploymentName={deployment_name}&api-version=2021-10-01"

# تحميل ملف Excel
excel_file = "Dental_Project_QnA.xlsx"  # حطي هنا مسار الملف الصحيح
df = pd.read_excel(excel_file)

# استخراج الأسئلة من العمود الصحيح
questions = df['Question'].tolist()

# هيدر الطلب
headers = {
    "Ocp-Apim-Subscription-Key": subscription_key,
    "Content-Type": "application/json"
}

# إرسال الأسئلة واستقبال الردود
for question in questions:
    data = {
        "question": question,
        "top": 1  # نجيب أفضل إجابة واحدة
    }

    response = requests.post(url, headers=headers, json=data)

    print(f"\nسؤال: {question}")
    print("Status Code:", response.status_code)

    if response.status_code == 200:
        answers = response.json().get("answers", [])
        if answers:
            print("الإجابة:", answers[0]["answer"])
        else:
            print("لا توجد إجابة.")
    else:
        print("فشل الاتصال - الرد:", response.text)
```
